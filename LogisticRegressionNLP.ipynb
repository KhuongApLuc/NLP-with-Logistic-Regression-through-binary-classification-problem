```python
#xử lý các giá trị NaN
import pandas as pd
df = pd.read_csv('final_describe.csv')
df.fillna('', inplace=True)
```

```python
df
```

#### Ghép cột description và user_tags, vector hoá dữ liệu

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Kết hợp cột description và tag thành một cột duy nhất
df['text'] = df['description'] + ' ' + df['user_tags']

# Tách dữ liệu thành training và testing set
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

# Chuyển văn bản thành vector
vectorizer = TfidfVectorizer(max_features=10000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Khởi tạo mô hình Logistic Regression
model = LogisticRegression()

# Huấn luyện mô hình
model.fit(X_train_tfidf, y_train)

# Dự đoán trên tập test
y_pred = model.predict(X_test_tfidf)

# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-score: {f1}')
```
